```python
# Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру
def name_doc():
  number = input("Введите номер документа: ")
  for data in documents:
    if data.get("number") == number:
      return (f'Владелец документа: {data["name"]}')
  return ('Документ не найден в базе')
```

```python
# Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
def shelf():
  number = input("Введите номер документа: ")
  for num in directories:
    if number in directories.get(num):
      return (f'Документ хранится на полке: {num}')
  return "Документ не найден в базе"
```

```python
# Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам
def lis():
  for shelf in directories:
    for num in directories[shelf]:
      found = next(e for e in documents if e['number'] == num)
      print(f'№: {num}, тип: {found["type"]}, владелец: {found["name"]}, полка хранения: {shelf}')
```

```python
# Пункт 4. Пользователь по команде "ads" может добавить новую полку
def ads():
  shelf = input("Введите номер полки: ")
  for direct in directories.items():
    if shelf in direct[0]:
      return 'Такая полка уже существует. Текущий перечень полок:' + ', ' .join(directories.keys())
      break
    else:
      directories[shelf] = []
      return 'Полка добавлена. Текущий перечень полок:' + ', ' .join(directories.keys())
```

```python

# Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
def rem_shelf():
  shelf = input("Введите номер полки: ")
  if len(directories[shelf]) == 0:
      del directories[shelf]
      return 'Полка удалена. Текущий перечень полок: ' + ', '.join(directories.keys())
  elif len(directories[shelf]) != 0:
      return 'На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: ' + ', '.join(directories.keys())
  else:
      return 'Такой полки не существует. Текущий перечень полок:' + ', '.join(directories.keys())
```

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
while True:
  print('ведите команду: p-узнать владельца документа по его номеру,s-по номеру документа узнать на какой полке он хранится,l-узнать полную информацию по всем документам,ads-добавить новую полку,ds-удалить существующую полку из данных')
  command = input('Введите команду: ')
  if command == 'p':
    print(name_doc())
  elif command == 's':
    print(shelf())
  elif command == 'l':
    print(lis())
  elif command == 'a':
    print(ads())
  elif command == 'ds':
    print(rem_shelf())
  elif command == 'q':
    print('Завершение программы')
    break
  print(documents)
  print(directories)
```
