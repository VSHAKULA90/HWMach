объединение таблиц | проверка

```python
df_big = pd.read_sql(
"""
SELECT
  *
FROM
  country_language cl
INNER JOIN
  country c
ON
  cl.country_code = c.code
WHERE
  is_official = 'True'
""",
"postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db"
)
df_big.head()
```

проверка объединеной таблицы df_big на выгрузку данных по полю is_official только True

```python
df_big['is_official'].value_counts()
```

выгрузка определеных полей из таблицы df_big | создание df_small

```python
df_small = df_big[['country_code','language','percentage','name','is_official']]
df_small.head()
```

проверка количества официальных языков
